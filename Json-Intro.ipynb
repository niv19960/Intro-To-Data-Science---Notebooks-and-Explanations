### Convert an object from Python into JSON 

We can Convert a Python object, into a JSON string by using the next method.

```python
# Create a Python object (dictionary):
x = {
  "name": "Ofir",
  "age": 23,
  "city": "Holon"
}
```

```python
# convert the dictionary into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(y)
print(type(y))
```

**Note** : We can convert Python objects of the following types, into JSON strings:
dict, list, tuple, string, int, float, True, False, None

Lets take a look for examples: 

```python
print(json.dumps({"name": "Niv", "age": 25})) #dict
print(json.dumps(["Apple", "Samsung"])) # list
print(json.dumps(("cup", "spoon"))) #tuple
print(json.dumps("hello world")) #string
print(json.dumps(550)) # int
print(json.dumps(51.87)) # float
print(json.dumps(True)) # True
print(json.dumps(False)) # False
print(json.dumps(None)) # None

#lets Convert a Python object containing all the legal data types:
x = {
  "name": "Eden",
  "age": 40,
  "married": True,
  "divorced": False,
  "children": ("Sivan","Yaron"),
  "pets": None,
  "cars": [
    {"model": "Kia", "mpg": 27.5},
    {"model": "Tesla", "mpg": 24.1}
  ]
}

print(json.dumps(x))
```

The json.dumps() method has parameters to order the keys in the result:

```python
json.dumps(x,sort_keys=True)
```

### Convert a JSON string to Python object

Lets create a Json: 

```python
jstring='{"id":"888", "name": "Ofir", "Class":"third year"}'
print(type(jstring))
```

```python
# Convert json into Python dict 
student_dict = json.loads(jstring) 
print(student_dict) 
print(type(student_dict))
print(student_dict['name'])
```

### Writing Json data into a file 

```python
# Creating a dict
data = {}
data['people'] = []
data['people'].append({
    'name': 'Scott',
    'website': 'stackabuse.com',
    'from': 'Nebraska'
})
data['people'].append({
    'name': 'Larry',
    'website': 'google.com',
    'from': 'Michigan'
})
data['people'].append({
    'name': 'Tim',
    'website': 'apple.com',
    'from': 'Alabama'
})

data
```

```python
# Writing Json data into a file 
with open('data.txt', 'w') as outfile:
    json.dump(data, outfile)
```

 ### Reading JSON from a File

```python
# Reading data back
with open('data.txt') as json_file:
    data = json.load(json_file)
    for p in data['people']:
        print('Name: ' + p['name'])
        print('Website: ' + p['website'])
        print('From: ' + p['from'])
        print('')
```
